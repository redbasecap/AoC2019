```python
#working but much to slow O^2

import time
with open ("input1.txt") as wires:
    wire1 = wires.readline().split(",")
    wire2 = wires.readline().split(",")


#First of all we want to decide which coordinate system we want to use:
#I choose the cartesian coordinate system
# y
# ^
# |
# |
# |
# |--------> x

#Testing Values
#wire1 = ["R75","D30","R83","U83","L12","D49","R71","U7","L72"]
#wire2 = ["U62","R66","U55","R34","D71","R55","D58","R83"]



#Parameter: the wire-movement_commands
def calc_min_distance(wire1, wire2):    

    #Class to create a dot in a coordinate system
    class Point:
        def __init__(self,X,Y):
            self.x = X
            self.y = Y
            self.counter = 0
            
        def set_to(self,X,Y,COUNTER):
            self.x = X
            self.y = Y
            self.counter = COUNTER

        def move(self,dx,dy):
            self.x = self.x+dx
            self.y = self.y+dy   
            self.counter += 1
    
        def get_counter(self):
            return(self.counter)
    
        def get_x(self):
            return(self.x)

        def get_y(self):
            return(self.y)

        def get_coordinates(self):
            return (self.x,self.y,self.counter)

        def __str__(self):
            return "Point(%s,%s)"%(self.x, self.y) 
        
    start = time.clock()
    
    start_point  = Point(0,0)
    moving_point = Point(0,0)
    #List of tuple to save all visited dots of the wire 
    coordinates = {1:[],2:[],"Intersections":[]}
    wire_counter = 1
    
    for wire in [wire1, wire2]:
        #reset the point to the 0 point of the coordinate system
        moving_point.set_to(0,0,0)
          
        for command in wire:
            #switch case
            if command[0] == 'U':
                for i in range(0,int(command[1::])):
                    moving_point.move(0,1)
                    coordinates[wire_counter].append(moving_point.get_coordinates())
            elif command[0] == 'D':
                for i in range(0,int(command[1::])):
                    moving_point.move(0,-1)
                    coordinates[wire_counter].append(moving_point.get_coordinates())
            elif command[0] == 'L':
                for i in range(0,int(command[1::])):
                    moving_point.move(-1,0)
                    coordinates[wire_counter].append(moving_point.get_coordinates())
            elif command[0] == 'R':
                for i in range(0,int(command[1::])):
                    moving_point.move(1,0)
                    coordinates[wire_counter].append(moving_point.get_coordinates())
        wire_counter += 1
        
        #Compare the every tuple (x,y,c) with each other and then add the both counter from the Point Class toghether
        #to get the length of the path
    print(len(coordinates[1]))    
    
    for k in range(0,len(coordinates[1])):
        for i in range(0,len(coordinates[2])):
            if (coordinates[1][k][0] == coordinates[2][i][0]) & (coordinates[1][k][1] == coordinates[2][i][1]):
                coordinates["Intersections"].append((coordinates[1][k][2]+coordinates[2][i][2]))
    
    ende = (time.clock()-start)
    print(ende)
    return min(coordinates["Intersections"])



calc_min_distance(wire1,wire2)
```

```python
#Day 3 Part 2
#Better Solution thx to https://github.com/sarahfossheim
with open ("input1.txt") as wires:
    wire1 = wires.readline().split(",")
    wire2 = wires.readline().split(",")


#First of all we want to decide which coordinate system we want to use:
#I choose the cartesian coordinate system
# y
# ^
# |
# |
# |
# |--------> x

#Testing Values
#wire1 = ["R75","D30","R83","U83","L12","D49","R71","U7","L72"]
#wire2 = ["U62","R66","U55","R34","D71","R55","D58","R83"]

#Parameter: the wire-movement_commands
def populate_grid(wire):    
    
    coordinates_all = {}

    # Set the current position
    x = 0
    y = 0
    step = 0
    
    for move in wire:
        direction = move[0]
        distance = int(move[1:])

        move_x = move_y = 0

        if direction == "L":
            move_x = -1
        if direction == "R":
            move_x = 1
        if direction == "D":
            move_y = -1
        if direction == "U":
            move_y = 1
            
        for _ in range(0,distance):
            x += move_x 
            y += move_y
            
            step += 1
            
            #(x,y) is the key and step the value
            if (x,y) not in coordinates_all:
                coordinates_all[(x,y)] = step
        
    return coordinates_all

    
line_1 = populate_grid(wire1)
line_2 = populate_grid(wire2)
intersections_keys = list(set(line_1.keys()) & set(line_2.keys()))
combined_steps = []
#search for the smallest value and return the key
for i in intersections_keys:
    combined_steps.append(line_1[i] + line_2[i])
    
print("Solution: ", min(combined_steps))
```
