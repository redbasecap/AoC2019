```python
with open ("input1.txt") as wires:
    wire1 = wires.readline().split(",")
    wire2 = wires.readline().split(",")


#First of all we want to decide which coordinate system we want to use:
#I choose the cartesian coordinate system
# y
# ^
# |
# |
# |
# |--------> x

#Testing Values
#wire1 = ["R10", "U20", "D30", "L33"]
#wire2 = ["R5", "R20", "D30", "U33"]


#Parameter: the wire-movement_commands
def build_grid(wire1, wire2):    
    
    #Class to create a dot in a coordinate system
    class Point:
        def __init__(self,X,Y):
            self.x = X
            self.y = Y
            
        def set_to(self,X,Y):
            self.x = X
            self.y = Y

        def move(self,dx,dy):
            self.x = self.x+dx
            self.y = self.y+dy            

        def get_x(self):
            return(self.x)

        def get_y(self):
            return(self.y)

        def get_coordinates(self):
            return (self.x,self.y)

        def __str__(self):
            return "Point(%s,%s)"%(self.x, self.y) 
    
    start_point  = Point(0,0)
    moving_point = Point(0,0)
    #List of tuple to save all visited dots of the wire 
    coordinates = {1:[],2:[],"Intersections":[]}
    counter = 1
    
    for wire in [wire1, wire2]:
        #reset the point to the 0 point of the coordinate system
        moving_point.set_to(0,0)
          
        for command in wire:
            #switch case
            if command[0] == 'U':
                for i in range(0,int(command[1::])):
                    moving_point.move(0,1)
                    coordinates[counter].append(moving_point.get_coordinates())
            elif command[0] == 'D':
                for i in range(0,int(command[1::])):
                    moving_point.move(0,-1)
                    coordinates[counter].append(moving_point.get_coordinates())
            elif command[0] == 'L':
                for i in range(0,int(command[1::])):
                    moving_point.move(-1,0)
                    coordinates[counter].append(moving_point.get_coordinates())
            elif command[0] == 'R':
                for i in range(0,int(command[1::])):
                    moving_point.move(1,0)
                    coordinates[counter].append(moving_point.get_coordinates())
            
        counter += 1
    
    #Find the intersection points
    coordinates["Intersections"] = set(coordinates[1]) & set(coordinates[2])
    return coordinates["Intersections"]
    
def manhatten_dist(x0,y0,x1,y1):
    return abs(x0-x1)+abs(y0-y1)

    
result = 10000

for [x,y] in build_grid(wire1,wire2):
    temp = manhatten_dist(0,0,x,y)
    if result > temp:
        result = temp
    
print(result)
```

```python
#Better Solution thx to https://github.com/sarahfossheim
with open ("input1.txt") as wires:
    wire1 = wires.readline().split(",")
    wire2 = wires.readline().split(",")


#First of all we want to decide which coordinate system we want to use:
#I choose the cartesian coordinate system
# y
# ^
# |
# |
# |
# |--------> x

#Testing Values
#wire1 = ["R10", "U20", "D30", "L33"]
#wire2 = ["R5", "R20", "D30", "U33"]


#Parameter: the wire-movement_commands
def populate_grid(wire):    
    
    coordinates_all = []

    # Set the current position
    x = 0
    y = 0
    
    for move in wire:
        direction = move[0]
        distance = int(move[1:])

        move_x = move_y = 0

        if direction == "L":
            move_x = -1
        if direction == "R":
            move_x = 1
        if direction == "D":
            move_y = -1
        if direction == "U":
            move_y = 1
            
        for _ in range(0,distance):
            x += move_x 
            y += move_y
            
            new_coordinates = (x,y)
            coordinates_all.append(new_coordinates)
        
    return coordinates_all
    
def manhatten_dist(x0,y0,x1,y1):
    return abs(x0-x1)+abs(y0-y1)

    
line_1 = populate_grid(wire1)
line_2 = populate_grid(wire2)
dist = []
intersections = list(set(line_1) & set(line_2))
for x,y in intersections:
    dist.append(manhatten_dist(0,0,x,y))

print("Solution: ",min(dist))
```
